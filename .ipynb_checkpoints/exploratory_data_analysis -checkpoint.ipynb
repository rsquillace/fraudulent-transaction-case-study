```python
trans_ip = fraud['ip_address'].values
ip_low = ip['lower_bound_ip_address'].values
ip_high = ip['upper_bound_ip_address'].values

i, j = np.where((trans_ip[:, None] >= ip_low) & (trans_ip[:, None] <= ip_high))
pd.DataFrame(
    np.column_stack([fraud.values[i], ip.values[j]]),
    columns=fraud.columns.append(ip.columns)
)
```
